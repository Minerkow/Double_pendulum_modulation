# Математические выкладки

В качестве обобщенных координат выберем углы $\phi_1, \phi_2$. Кинетическая
энергия:
$$
    T = \frac{1}{2}(m_a v_a^2 + m_b v_b^2).
$$
В проекциях на оси координат:
$$
    v_{ax} = a \dot{\phi_1} \cos{\phi_1}, 
    \\
    v_{ay} = a \dot{\phi_1} \sin{\phi_1}. 
$$
Для точки B имеем:
$$
    v_{bx} = a \dot{\phi_1} \cos{\phi_1} + b \dot{\phi_2} \cos{\phi_2}, 
    \\
    v_{by} = a \dot{\phi_1} \sin{\phi_1} + b \dot{\phi_2} \sin{\phi_2}.
$$
Потенциальная энергия сил тяжести:
$$
    \Pi = m_a g a \cos{\phi_1} - m_b g(a\cos{\phi_1} + b \cos{\phi_2}).
$$
Тогда функция Лагранжа принимает вид:
$$
    L = T - \Pi = \frac{1}{2}m_a a^2 \dot{\phi_1^2} + \frac{1}{2}m_b(a^2 \dot{\phi_1^2} + b^2 \dot{\phi_2^2} + 2ab\dot{\phi_1}\dot{\phi_2}\cos{(\phi_1 - \phi_2}))+m_aga\cos{\phi_1} + m_bg(a\cos{\phi_1} + b\cos{\phi_2}).
$$
Запишем уравнения Лагранжа в матричной форме с учетом дисипативной силы:
$$
    \begin{pmatrix}
        (m_a+m_b)a^2 & m_bab\cos{(\phi_1 - \phi_2)} \\
         m_bab\cos{(\phi_1 - \phi_2)} & m_bb^2
    \end{pmatrix}
    \begin{pmatrix}
        \ddot{\phi_1} \\
        \ddot{\phi_2}
    \end{pmatrix}
    +
    \begin{pmatrix}
        \beta & 0 \\
        0 & \beta
    \end{pmatrix}
    \begin{pmatrix}
        \dot{\phi_1} \\
        \dot{\phi_2}
    \end{pmatrix}
    +
    \begin{pmatrix}
            ga(m_a + m_b) & 0 \\
            0 & gbm_b
    \end{pmatrix}
    \begin{pmatrix}
        \sin{\phi_1} \\
        \sin{\phi_2} 
    \end{pmatrix}
    -m_bab\sin{(\phi_2 - \phi_1)}
    \begin{pmatrix}
         \dot{\phi_2}^2 \\
         \dot{\phi_1}^2
    \end{pmatrix}
    = 0
$$
Для малых углов получим:
$$
    \begin{pmatrix}
        (m_a + m_b)a^2 & m_bab \\
        m_bab & m_bb^2
    \end{pmatrix}
        \begin{pmatrix}
        \ddot{\phi_1} \\
        \ddot{\phi_2}
    \end{pmatrix}
    +
        \begin{pmatrix}
        \beta & 0 \\
        0 & \beta
    \end{pmatrix}
    \begin{pmatrix}
        \dot{\phi_1} \\
        \dot{\phi_2}
    \end{pmatrix}
    + 
    \begin{pmatrix}
            ga(m_a + m_b) & 0 \\
            0 & gbm_b
    \end{pmatrix}
    \begin{pmatrix}
        \phi_1 \\
        \phi_2
    \end{pmatrix}
    =0
$$
